```python
# Import the required libraries
import hashlib
import time
import datetime
import random
import matplotlib.pyplot as plt
```

```python
class Block:
    """A simplified block in the blockchain."""
    def __init__(self, index, timestamp, data, previous_hash):
        self.index = index
        self.timestamp = timestamp
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        """Generates the hash for the block."""
        block_string = str(self.index) + str(self.timestamp) + str(self.data) + str(self.previous_hash)
        return hashlib.sha256(block_string.encode()).hexdigest()

class Blockchain:
    """A simplified blockchain."""
    def __init__(self):
        self.chain = [self.create_genesis_block()]

    def create_genesis_block(self):
        """Creates the first block in the blockchain."""
        return Block(0, datetime.datetime.now(), "Genesis Block", "0")

    def get_latest_block(self):
        """Returns the most recently added block."""
        return self.chain[-1]

    def add_block(self, new_block):
        """Adds a new block to the chain."""
        new_block.previous_hash = self.get_latest_block().hash
        new_block.hash = new_block.calculate_hash()
        # Simulate network delay and consensus by introducing a sleep timer
        time.sleep(0.1) # Simulates network propagation and validation
        self.chain.append(new_block)
```

```python
# Create a new blockchain instance
cloud_access_chain = Blockchain()

# Simulation parameters
num_users = 100
authentication_times = []

# Simulate authentication for each user
print("Starting simulation for 100 users...")
for user_id in range(1, num_users + 1):
    # Simulate a cloud access request from a user
    user_data = f"User {user_id} accessing cloud at {datetime.datetime.now()}"

    # Measure the time to process the authentication transaction
    start_time = time.perf_counter()
    new_block = Block(
        index=len(cloud_access_chain.chain),
        timestamp=datetime.datetime.now(),
        data=user_data,
        previous_hash=""
    )
    cloud_access_chain.add_block(new_block)
    end_time = time.perf_counter()

    # Calculate and store the authentication latency
    latency = (end_time - start_time) * 1000  # Convert to milliseconds
    authentication_times.append(latency)
    print(f"User {user_id}: Authentication Latency = {latency:.2f} ms")

# Calculate the average authentication latency
average_latency = sum(authentication_times) / num_users
print(f"\nAverage Authentication Latency for {num_users} users: {average_latency:.2f} ms")
```

```python
# Plot the results using Matplotlib
plt.figure(figsize=(12, 6))
plt.plot(range(1, num_users + 1), authentication_times, marker='o', linestyle='-', color='b', label='Latency per User')
plt.axhline(y=average_latency, color='r', linestyle='--', label=f'Average Latency ({average_latency:.2f} ms)')

# Customize the plot
plt.title('Authentication Latency for 100 Users using Blockchain')
plt.xlabel('User ID')
plt.ylabel('Latency (ms)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
