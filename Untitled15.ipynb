```python
import bcrypt
import pyotp
import time
import matplotlib.pyplot as plt
import pandas as pd

# In-memory database for demonstration
users_db = {}

# --- User Registration ---
def register_user(username, password):
    """Hashes password with bcrypt and generates a TOTP secret."""
    salt = bcrypt.gensalt(rounds=12)  # High cost factor for security
    hashed_password = bcrypt.hashpw(password.encode('utf-8'), salt)
    otp_secret = pyotp.random_base32()
    users_db[username] = {
        'hashed_password': hashed_password,
        'otp_secret': otp_secret
    }
    print(f"User '{username}' registered. Save this secret for your authenticator app: {otp_secret}")

# --- User Login ---
def login_user(username, password, user_otp):
    """Authenticates a user with both password and OTP."""
    if username not in users_db:
        print("Error: User not found.")
        return False, None

    user_data = users_db[username]
    hashed_password = user_data['hashed_password']

    # Verify the password using bcrypt's slow comparison
    if not bcrypt.checkpw(password.encode('utf-8'), hashed_password):
        print("Error: Incorrect password.")
        return False, None

    # Verify the OTP using pyotp
    totp = pyotp.TOTP(user_data['otp_secret'])
    if not totp.verify(user_otp):
        print("Error: Incorrect OTP.")
        return False, None

    print(f"Success: User '{username}' logged in successfully!")
    return True, totp.now()

# --- Simulate and benchmark authentication latency ---
def benchmark_latency(password, cost_factors):
    """Measures and plots latency for different bcrypt cost factors."""
    registration_latencies = []
    login_latencies = []

    for rounds in cost_factors:
        # Measure registration time
        start_time = time.time()
        salt = bcrypt.gensalt(rounds=rounds)
        bcrypt.hashpw(password.encode('utf-8'), salt)
        registration_latencies.append(time.time() - start_time)

        # Measure login verification time (using the same cost factor)
        start_time = time.time()
        bcrypt.checkpw(password.encode('utf-8'), bcrypt.hashpw(password.encode('utf-8'), salt))
        login_latencies.append(time.time() - start_time)

    df = pd.DataFrame({
        'Cost Factor (Rounds)': cost_factors,
        'Registration Latency (s)': registration_latencies,
        'Login Latency (s)': login_latencies
    })

    # Plotting the results using matplotlib
    plt.figure(figsize=(10, 6))
    plt.plot(df['Cost Factor (Rounds)'], df['Login Latency (s)'], label='Login Latency', marker='o')
    plt.plot(df['Cost Factor (Rounds)'], df['Registration Latency (s)'], label='Registration Latency', marker='o')
    plt.xlabel('Bcrypt Cost Factor (Rounds)')
    plt.ylabel('Latency (Seconds)')
    plt.title('Bcrypt Hashing Latency vs. Cost Factor')
    plt.legend()
    plt.grid(True)
    plt.show()

    return df

# --- Execution ---
if __name__ == "__main__":
    # Example usage
    username = "test_user"
    password = "MySecurePassword123"
    register_user(username, password)

    # To get the OTP, you can use the secret from the registration printout with an authenticator app,
    # or generate one manually for this demonstration.
    secret = users_db[username]['otp_secret']
    totp = pyotp.TOTP(secret)
    current_otp = totp.now()
    print(f"Generated current OTP: {current_otp}")

    # Test a successful login
    print("\n--- Testing Successful Login ---")
    login_user(username, password, current_otp)

    # Test a failed login (bad password)
    print("\n--- Testing Failed Login (Bad Password) ---")
    login_user(username, "WrongPassword", current_otp)

    # Test a failed login (bad OTP)
    print("\n--- Testing Failed Login (Bad OTP) ---")
    login_user(username, password, "000000") # A fake OTP

    # Run latency benchmark
    print("\n--- Benchmarking Bcrypt Latency ---")
    cost_factors_to_test = [8, 10, 12, 14, 16]
    latency_results = benchmark_latency(password, cost_factors_to_test)
    print("Latency Results:")
    print(latency_results)
```
