```python
import numpy as np
import matplotlib.pyplot as plt
import time
import threading
import random
```

```python
def authenticate_user(user_id, latencies, lock):
    """
    Simulates a single user's authentication process.
    """
    # Simulate network and server latency with a random delay (e.g., 50ms to 500ms)
    delay_ms = random.uniform(50, 500)

    start_time = time.time()
    time.sleep(delay_ms / 1000)  # Convert milliseconds to seconds
    end_time = time.time()

    latency = (end_time - start_time) * 1000  # Convert back to milliseconds

    # Use a lock to safely append latency to the shared list
    with lock:
        latencies.append(latency)

    print(f"User {user_id} authenticated in {latency:.2f} ms")
```

```python
from concurrent.futures import ThreadPoolExecutor

NUM_USERS = 100
latencies = []
lock = threading.Lock()

def run_simulation(num_users):
    """
    Manages the execution of concurrent authentication requests.
    """
    with ThreadPoolExecutor(max_workers=num_users) as executor:
        for i in range(1, num_users + 1):
            executor.submit(authenticate_user, i, latencies, lock)

print(f"Starting authentication simulation for {NUM_USERS} users...")
run_simulation(NUM_USERS)
print("\nSimulation complete.")
```

```python
# Create the Matplotlib plot in a new cell

# Use the `%matplotlib inline` magic command for Colab
%matplotlib inline

# Create a histogram of the latency data
plt.figure(figsize=(12, 7))
plt.hist(latencies, bins=30, edgecolor='black', alpha=0.7)

# Add titles and labels for clarity
plt.title('Authentication Latency Distribution for 100 Cloud Users', fontsize=16)
plt.xlabel('Latency (milliseconds)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add key statistics as text on the plot
mean_latency = np.mean(latencies)
median_latency = np.median(latencies)
plt.axvline(mean_latency, color='red', linestyle='dashed', linewidth=2, label=f'Mean Latency: {mean_latency:.2f} ms')
plt.axvline(median_latency, color='green', linestyle='dashed', linewidth=2, label=f'Median Latency: {median_latency:.2f} ms')
plt.legend()

plt.show()

# Print summary statistics
print(f"Mean Latency: {mean_latency:.2f} ms")
print(f"Median Latency: {median_latency:.2f} ms")
print(f"Max Latency: {np.max(latencies):.2f} ms")
print(f"Min Latency: {np.min(latencies):.2f} ms")
```
