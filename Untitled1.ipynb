```python
import time
import bcrypt
import scrypt
from argon2 import PasswordHasher
import matplotlib.pyplot as plt

# --- Configuration ---
PASSWORD = b"mysecretpassword"
ROUNDS_B = 12 # Bcrypt rounds (higher is slower/more secure)
N_SCRYPT = 16384 # Scrypt N (CPU/Memory cost parameter, power of 2)
R_SCRYPT = 8    # Scrypt r (Blocksize)
P_SCRYPT = 1    # Scrypt p (Parallelization parameter)
MEM_ARGON = 65536 # Argon2 memory cost (KiB)
TIME_ARGON = 3    # Argon2 time cost
PARA_ARGON = 4    # Argon2 parallelization

# --- Performance Measurement Functions ---

def time_bcrypt():
    start_time = time.time()
    bcrypt.hashpw(PASSWORD, bcrypt.gensalt(rounds=ROUNDS_B))
    end_time = time.time()
    return end_time - start_time

def time_scrypt():
    start_time = time.time()
    # Corrected parameter from 'dklen' to 'buflen' based on scrypt library documentation
    scrypt.hash(PASSWORD, salt=b"somesalt", N=N_SCRYPT, r=R_SCRYPT, p=P_SCRYPT, buflen=32)
    end_time = time.time()
    return end_time - start_time

def time_argon2():
    ph = PasswordHasher(memory_cost=MEM_ARGON, time_cost=TIME_ARGON, parallelism=PARA_ARGON)
    start_time = time.time()
    ph.hash(PASSWORD.decode('utf-8'))
    end_time = time.time()
    return end_time - start_time

# --- Run Tests ---
print("Running performance tests...")

# Bcrypt
bcrypt_time = time_bcrypt()
print(f"Bcrypt Time: {bcrypt_time:.4f} seconds")

# Scrypt
scrypt_time = time_scrypt()
print(f"Scrypt Time: {scrypt_time:.4f} seconds")

# Argon2
argon2_time = time_argon2()
print(f"Argon2 Time: {argon2_time:.4f} seconds")

# --- Visualize Results with Matplotlib ---
algorithms = ['Bcrypt', 'Scrypt', 'Argon2']
times = [bcrypt_time, scrypt_time, argon2_time]

plt.figure(figsize=(8, 5))
plt.bar(algorithms, times, color=['blue', 'green', 'red'])
plt.ylabel('Time (seconds)')
plt.title('Password Hashing Performance Comparison')
plt.grid(axis='y', linestyle='--')
plt.show()
```
