```python
import bcrypt

def hash_password(password):
    """Hashes a password using a randomly generated salt."""
    # Generate a salt and hash the password
    salt = bcrypt.gensalt(rounds=12) # Use a high work factor for security
    hashed = bcrypt.hashpw(password.encode('utf-8'), salt)
    return hashed

def check_password(password, hashed_password):
    """Verifies a password against a stored hash."""
    return bcrypt.checkpw(password.encode('utf-8'), hashed_password)

# Example usage of bcrypt
password = "mySecretPassword123"
print(f"Original password: {password}")

# Store the hashed password in a simulated database
stored_hash = hash_password(password)
print(f"Stored hashed password: {stored_hash}")

# Simulate a correct login attempt
if check_password(password, stored_hash):
    print("Login successful with correct password!")
else:
    print("Login failed with incorrect password.")

# Simulate an incorrect login attempt
incorrect_password = "wrongPassword123"
if check_password(incorrect_password, stored_hash):
    print("Login successful with incorrect password.")
else:
    print("Login failed with incorrect password. Defense against stolen credentials is working.")
```

```python
import pyotp
import time

def generate_totp_secret():
    """Generates a random base32 secret key."""
    return pyotp.random_base32()

def get_totp_uri(username, secret):
    """Generates a URI for provisioning an authenticator app."""
    return pyotp.totp.TOTP(secret).provisioning_uri(
        name=username,
        issuer_name="MySecurityApp"
    )

def verify_totp(secret, user_input_otp):
    """Verifies the user-provided OTP."""
    totp = pyotp.TOTP(secret)
    return totp.verify(user_input_otp)

# Example usage of OTP
secret_key = generate_totp_secret()
user_id = "user123"

# Print the provisioning URI to set up an authenticator app
print(f"\nScan this QR code with your authenticator app:")
print(get_totp_uri(user_id, secret_key))
print("You would typically generate a QR code from this URI.")

# Simulate user entering an OTP
current_otp = pyotp.TOTP(secret_key).now()
print(f"Simulated valid OTP (for demonstration): {current_otp}")

# Verification
user_input = current_otp # Correct OTP
if verify_totp(secret_key, user_input):
    print("OTP verification successful!")
else:
    print("OTP verification failed.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data representing defense against different attack types
attack_types = ['Credential Stuffing', 'Brute-force (Offline)', 'Phishing', 'Man-in-the-Middle']

# Simulated effectiveness scores (0-10)
# Higher is more effective
bcrypt_effectiveness = [0, 9, 2, 1]  # Bcrypt helps little with attacks that steal a valid session
otp_effectiveness = [10, 0, 8, 9]     # OTP prevents unauthorized access even if the password is stolen
combined_effectiveness = np.array(bcrypt_effectiveness) + np.array(otp_effectiveness)

x = np.arange(len(attack_types))
width = 0.25

fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars for bcrypt, OTP, and combined effectiveness
ax.bar(x - width, bcrypt_effectiveness, width, label='Bcrypt (Password Hashing)', color='#6a8e32')
ax.bar(x, otp_effectiveness, width, label='OTP (Multi-factor Auth)', color='#2e578c')
ax.bar(x + width, combined_effectiveness, width, label='Combined (MFA)', color='#d4763b')

# Set chart details
ax.set_ylabel('Defense Effectiveness (Relative Score)')
ax.set_title('Defense Against Cyber Attacks: Bcrypt vs. OTP vs. Combined')
ax.set_xticks(x)
ax.set_xticklabels(attack_types)
ax.set_ylim(0, 15)
ax.legend()

# Add explanations for each defense score
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(ax.patches[:len(attack_types)])
add_labels(ax.patches[len(attack_types):len(attack_types)*2])
add_labels(ax.patches[len(attack_types)*2:])

# Add descriptive text below the plot
plt.figtext(0.5, -0.05,
            """
            **Analysis:**
            - **Credential Stuffing:** The combined approach is highly effective. Even if an attacker has stolen credentials from a data breach, they cannot log in without the one-time password.
            - **Brute-force (Offline):** Bcrypt excels here by making it computationally expensive to crack passwords, but OTP is not involved since the attacker is not trying to log in via an interface.
            - **Phishing/Man-in-the-Middle:** While bcrypt is irrelevant once credentials are stolen, the OTP acts as a secondary verification, greatly diminishing the value of a compromised password.
            """,
            ha='center', fontsize=10, style='italic',
            bbox={"facecolor":"white", "alpha":0.5, "pad":5})

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```
