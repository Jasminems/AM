```python
    import bcrypt

    def hash_password_bcrypt(password):
        hashed = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt())
        return hashed

    def verify_password_bcrypt(password, hashed_password):
        return bcrypt.checkpw(password.encode('utf-8'), hashed_password)
```

```python
    import hashlib

    def hash_password_simple(password):
        return hashlib.sha256(password.encode('utf-8')).hexdigest()

    def verify_password_simple(password, hashed_password):
        return hashlib.sha256(password.encode('utf-8')).hexdigest() == hashed_password
```

```python
    import time

    def simulate_brute_force(target_hash, hash_function, possible_passwords):
        start_time = time.time()
        for attempt in possible_passwords:
            if hash_function(attempt) == target_hash:
                return time.time() - start_time, True
        return time.time() - start_time, False

    # Example usage:
    password = "password123"
    bcrypt_hash = hash_password_bcrypt(password)
    simple_hash = hash_password_simple(password)

    possible_passwords = [f"password{i}" for i in range(1000)] # Simplified dictionary attack

    bcrypt_time, _ = simulate_brute_force(bcrypt_hash, lambda p: hash_password_bcrypt(p), possible_passwords)
    simple_time, _ = simulate_brute_force(simple_hash, hash_password_simple, possible_passwords)
```

```python
    import matplotlib.pyplot as plt

    hashing_methods = ['Bcrypt', 'Simple Hash (SHA-256)']
    cracking_times = [bcrypt_time, simple_time] # Use actual measured times

    plt.figure(figsize=(8, 6))
    plt.bar(hashing_methods, cracking_times, color=['skyblue', 'lightcoral'])
    plt.ylabel('Time to Crack (seconds)')
    plt.title('Simulated Brute-Force Attack Cracking Time')
    plt.show()
```

```python
    attack_types = ['Brute-Force', 'Rainbow Table']
    bcrypt_vulnerability = [0.2, 0.1] # Lower vulnerability for bcrypt
    simple_hash_vulnerability = [0.8, 0.9] # Higher vulnerability for simple hash

    x = range(len(attack_types))
    plt.figure(figsize=(8, 6))
    plt.bar([i - 0.2 for i in x], bcrypt_vulnerability, width=0.4, label='Bcrypt', color='skyblue')
    plt.bar([i + 0.2 for i in x], simple_hash_vulnerability, width=0.4, label='Simple Hash', color='lightcoral')
    plt.xticks(x, attack_types)
    plt.ylabel('Relative Vulnerability Score')
    plt.title('Comparison of Hashing Algorithm Vulnerabilities to Cyberattacks')
    plt.legend()
    plt.show()
```
