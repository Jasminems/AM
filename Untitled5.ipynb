```python
import bcrypt
import time
import matplotlib.pyplot as plt
import numpy as np

def benchmark_bcrypt(password, rounds):
    """Hashes and verifies a password with bcrypt for a given number of rounds, measuring time."""
    start_time = time.time()
    hashed_password = bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt(rounds))
    end_hash_time = time.time()
    bcrypt.checkpw(password.encode('utf-8'), hashed_password)
    end_verify_time = time.time()
    return (end_hash_time - start_time, end_verify_time - end_hash_time)

# Define parameters for the benchmark
password = "my_secure_password_123"
bcrypt_rounds = [8, 10, 12, 14]
hash_times = []
verify_times = []

# Run the bcrypt benchmark
print("Benchmarking bcrypt performance...")
for rounds in bcrypt_rounds:
    hash_time, verify_time = benchmark_bcrypt(password, rounds)
    hash_times.append(hash_time)
    verify_times.append(verify_time)
    print(f"Rounds: {rounds} | Hashing Time: {hash_time:.4f}s | Verification Time: {verify_time:.4f}s")
```

```python
import pyotp
import qrcode
from IPython.display import Image, display

# Generate a random secret key for the user
# This key should be stored securely in a real application
secret = pyotp.random_base32()

# Create a TOTP object
totp = pyotp.TOTP(secret)

def generate_qr_code(username, issuer_name, totp_secret):
    """Generates a QR code for authenticator apps."""
    uri = pyotp.totp.TOTP(totp_secret).provisioning_uri(
        name=username,
        issuer_name=issuer_name
    )
    img = qrcode.make(uri)
    img_path = "otp_qr_code.png"
    img.save(img_path)
    print(f"QR Code saved as {img_path}. Scan with your authenticator app.")
    display(Image(img_path))

def verify_otp_example():
    """Demonstrates how to generate and verify an OTP."""
    print(f"\nYour secret key is: {secret}")
    generate_qr_code("testuser@example.com", "Colab MFA Demo", secret)

    # Simulate OTP verification
    user_otp = input("Enter the 6-digit code from your authenticator app: ")

    # The verification is very fast, so timing is not a major security factor here
    is_valid = totp.verify(user_otp)
    if is_valid:
        print("OTP is valid. Multi-factor authentication successful!")
    else:
        print("OTP is invalid. Authentication failed.")

# Run the OTP example
verify_otp_example()
```

```python
# Comparison plotting
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35

x = np.arange(len(bcrypt_rounds))
ax.bar(x - width/2, hash_times, width, label='Bcrypt Hashing Time')
ax.bar(x + width/2, verify_times, width, label='Bcrypt Verification Time')

# OTP verification time is negligible in comparison
otp_verification_time = 0.0001
ax.axhline(y=otp_verification_time, color='r', linestyle='--', label=f'OTP Verification Time (approx {otp_verification_time:.4f}s)')

ax.set_title('Bcrypt Hashing vs. OTP Verification Performance')
ax.set_xlabel('Bcrypt Rounds (Work Factor)')
ax.set_ylabel('Time in seconds')
ax.set_xticks(x)
ax.set_xticklabels(bcrypt_rounds)
ax.legend()
plt.show()
```
