```python
import bcrypt
import pyotp
import time

# Function to register a new user
def register_user(username, password):
    # Generate a salt and hash the password with bcrypt
    salt = bcrypt.gensalt()
    hashed_password = bcrypt.hashpw(password.encode('utf-8'), salt)

    # Generate a TOTP secret key for the user
    otp_secret = pyotp.random_base32()

    # Simulate a user database
    return {'username': username, 'hashed_password': hashed_password, 'otp_secret': otp_secret}

# Function to perform the first-factor authentication (password)
def verify_password(user, password):
    return bcrypt.checkpw(password.encode('utf-8'), user['hashed_password'])

# Function to perform the second-factor authentication (OTP)
def verify_otp(user, otp_code):
    totp = pyotp.TOTP(user['otp_secret'])
    return totp.verify(otp_code)

# Function to simulate a full MFA login
def mfa_login(user, password, otp_code):
    if verify_password(user, password):
        print("Password verified. Proceeding to OTP check...")
        if verify_otp(user, otp_code):
            print("OTP verified. Login successful!")
            return True
        else:
            print("Invalid OTP. Login failed.")
            return False
    else:
        print("Invalid password. Login failed.")
        return False

# Example usage
# --- User registration ---
new_user = register_user("testuser", "StrongPassword123")
print(f"New user registered with OTP secret: {new_user['otp_secret']}")

# --- Generate OTP (simulating a user's authenticator app) ---
totp = pyotp.TOTP(new_user['otp_secret'])
current_otp = totp.now()
print(f"Current OTP for login: {current_otp}")

# --- Simulate a login attempt ---
print("\nAttempting successful login...")
mfa_login(new_user, "StrongPassword123", current_otp)

print("\nAttempting failed password login...")
mfa_login(new_user, "WrongPassword", current_otp)

print("\nAttempting failed OTP login...")
time.sleep(30) # Wait for the OTP to expire
mfa_login(new_user, "StrongPassword123", totp.now())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import bcrypt
import time

# --- Simulation 1: Brute-force attack on bcrypt vs a simple hash ---

def simulate_bcrypt_crack(cost_factor):
    """Simulates the time taken for a brute-force attack on bcrypt."""
    password = "test_password"
    salt = bcrypt.gensalt(rounds=cost_factor)
    hashed_password = bcrypt.hashpw(password.encode('utf-8'), salt)

    start_time = time.time()
    # Simulate a brute-force attempt (checking only a few passwords)
    # In reality, this would take an extremely long time.
    for i in range(100):
        bcrypt.checkpw(f"fake_password_{i}".encode('utf-8'), hashed_password)
    end_time = time.time()
    return (end_time - start_time) / 100 # Time per password check

# Define different cost factors
cost_factors = [8, 12, 16]
bcrypt_crack_times = [simulate_bcrypt_crack(cost) for cost in cost_factors]
simple_hash_time = 0.000001 # A simple hash is very fast

# --- Simulation 2: Vulnerability comparison with and without OTP ---

# Effectiveness of attacks against different configurations
# Scale: 0 (not effective) to 1 (fully effective)
attacks = ['Brute-Force', 'Man-in-the-Middle', 'Credential Stuffing']
security_features = ['Bcrypt Only', 'Bcrypt + OTP']

# Data representing the security effectiveness (higher is better)
# Brute-Force: OTP doesn't affect it, but bcrypt resists it.
# MitM: OTP is highly effective, password less so.
# Credential Stuffing: OTP is highly effective, password is not at all.
effectiveness_data = {
    'Bcrypt Only': [0.8, 0.5, 0.2], # Bcrypt resists brute-force, is vulnerable to MitM and Credential Stuffing.
    'Bcrypt + OTP': [0.8, 0.9, 0.9] # OTP mitigates MitM and Credential Stuffing.
}

# --- Plotting the results using matplotlib ---

# Plot 1: Brute-force resistance
plt.figure(figsize=(10, 6))
plt.bar(
    [f"Bcrypt (Cost={c})" for c in cost_factors],
    bcrypt_crack_times,
    color='skyblue',
    label='Bcrypt Resistance (Simulated)'
)
plt.axhline(y=simple_hash_time * 1e6, color='r', linestyle='--', label='Simple Hash (Extremely fast)')
plt.yscale('log') # Use log scale to show the massive difference
plt.title('Bcrypt Brute-Force Resistance vs. Cost Factor')
plt.xlabel('Algorithm')
plt.ylabel('Simulated Time per Password Check (Log Scale)')
plt.legend()
plt.grid(axis='y')
plt.show()

# Plot 2: Cyber attack comparison
x = np.arange(len(attacks))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width/2, effectiveness_data['Bcrypt Only'], width, label='Bcrypt Only', color='lightcoral')
rects2 = ax.bar(x + width/2, effectiveness_data['Bcrypt + OTP'], width, label='Bcrypt + OTP', color='teal')

ax.set_title('Security Effectiveness Against Cyber Attacks')
ax.set_xlabel('Cyber Attack Type')
ax.set_ylabel('Security Effectiveness (Higher is better)')
ax.set_xticks(x)
ax.set_xticklabels(attacks)
ax.set_ylim(0, 1.1)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.1f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```
