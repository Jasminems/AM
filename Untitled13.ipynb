```python
import hashlib
import json
import time

class Blockchain:
    def __init__(self):
        self.chain = []
        self.pending_transactions = []
        self.create_block(proof=1, previous_hash='0')

    def create_block(self, proof, previous_hash):
        block = {
            'index': len(self.chain) + 1,
            'timestamp': time.time(),
            'transactions': self.pending_transactions,
            'proof': proof,
            'previous_hash': previous_hash or self.hash(self.chain[-1])
        }
        self.pending_transactions = []
        self.chain.append(block)
        return block

    @staticmethod
    def hash(block):
        block_string = json.dumps(block, sort_keys=True).encode()
        return hashlib.sha256(block_string).hexdigest()

    def add_transaction(self, sender, recipient, data):
        self.pending_transactions.append({
            'sender': sender,
            'recipient': recipient,
            'data': data
        })
        return self.last_block['index'] + 1

    @property
    def last_block(self):
        return self.chain[-1]

    def proof_of_work(self, last_proof):
        proof = 0
        while self.valid_proof(last_proof, proof) is False:
            proof += 1
        return proof

    @staticmethod
    def valid_proof(last_proof, proof):
        guess = f'{last_proof}{proof}'.encode()
        guess_hash = hashlib.sha256(guess).hexdigest()
        return guess_hash[:4] == "0000"

# Initialize blockchain
blockchain = Blockchain()
```

```python
from Crypto.PublicKey import RSA
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256

def generate_key_pair():
    key = RSA.generate(2048)
    return key.export_key(), key.publickey().export_key()

def sign_data(private_key_pem, data):
    private_key = RSA.import_key(private_key_pem)
    h = SHA256.new(data.encode())
    signature = pkcs1_15.new(private_key).sign(h)
    return signature

def verify_signature(public_key_pem, data, signature):
    public_key = RSA.import_key(public_key_pem)
    h = SHA256.new(data.encode())
    try:
        pkcs1_15.new(public_key).verify(h, signature)
        return True
    except (ValueError, TypeError):
        return False
```

```python
import bcrypt

def hash_password(password):
    return bcrypt.hashpw(password.encode('utf-8'), bcrypt.gensalt())

def check_password(password, hashed_password):
    return bcrypt.checkpw(password.encode('utf-8'), hashed_password)
```

```python
def user_registration(username, password, blockchain, user_db):
    private_key, public_key = generate_key_pair()
    hashed_password = hash_password(password)

    # Simulate storing on blockchain and local DB
    user_db[username] = {
        'public_key': public_key,
        'hashed_password': hashed_password,
        'private_key': private_key # For simulation only, never store privately
    }

    # Store minimal, public info on the blockchain
    registration_data = f"User {username} registered with public key {public_key.decode()}"
    blockchain.add_transaction("System", username, registration_data)

    print(f"User {username} registered. Public key added to blockchain.")

def user_authentication(username, password, blockchain, user_db):
    if username not in user_db:
        return False, "User not found"

    user_info = user_db[username]
    # Factor 1: Password check using bcrypt
    if not check_password(password, user_info['hashed_password']):
        return False, "Incorrect password"

    # Factor 2: PKI-based signature
    challenge = f"Authenticate {username} at {time.time()}"

    # In a real system, the user signs this challenge via a client-side app
    signature = sign_data(user_info['private_key'], challenge)

    # The server verifies the signature against the public key from the blockchain
    is_verified = verify_signature(user_info['public_key'], challenge, signature)

    if is_verified:
        # Log successful authentication on the blockchain
        blockchain.add_transaction(username, "Cloud-Access", f"Authentication success for {username}.")
        return True, "Authentication successful"
    else:
        # Log failed authentication
        blockchain.add_transaction(username, "Cloud-Access", f"Authentication failed for {username}.")
        return False, "PKI signature failed"

def revoke_certificate(username, blockchain, user_db):
    if username in user_db:
        # Simulate storing revocation on blockchain
        revocation_data = f"Certificate for user {username} has been revoked."
        blockchain.add_transaction("Authority", username, revocation_data)
        print(f"Revocation transaction for {username} broadcasted to the blockchain.")
```

```python
import time
import matplotlib.pyplot as plt
import numpy as np

# Simulate a traditional centralized system
class CentralizedSystem:
    def __init__(self):
        self.user_db = {}
        self.pki_ca = {}

    def register(self, username, password):
        private_key, public_key = generate_key_pair()
        hashed_password = hash_password(password)
        self.user_db[username] = {'hashed_password': hashed_password, 'private_key': private_key}
        self.pki_ca[username] = public_key

    def authenticate(self, username, password):
        if username not in self.user_db: return False

        # Factor 1: Password
        if not check_password(password, self.user_db[username]['hashed_password']): return False

        # Factor 2: PKI Signature check
        challenge = f"Authenticate {username} at {time.time()}"
        signature = sign_data(self.user_db[username]['private_key'], challenge)
        return verify_signature(self.pki_ca[username], challenge, signature)

    def revoke(self, username):
        if username in self.pki_ca:
            del self.pki_ca[username]
            return True
        return False

# Performance simulation function
def run_performance_test(blockchain_auth_func, centralized_auth_func, num_users=100, num_auth_attempts=1000):
    # Setup
    print("Setting up simulations...")
    blockchain_sys = Blockchain()
    centralized_sys = CentralizedSystem()
    user_data = {}

    for i in range(num_users):
        username = f'user{i}'
        password = 'password123'
        user_registration(username, password, blockchain_sys, user_data)
        centralized_sys.register(username, password)

    # Measure latency
    print("Measuring authentication latency...")
    blockchain_latencies = []
    centralized_latencies = []

    for _ in range(num_auth_attempts):
        user_to_auth = f'user{np.random.randint(0, num_users)}'

        start_time = time.time()
        blockchain_auth_func(user_to_auth, 'password123', blockchain_sys, user_data)
        blockchain_latencies.append(time.time() - start_time)

        start_time = time.time()
        centralized_auth_func(user_to_auth, 'password123')
        centralized_latencies.append(time.time() - start_time)

    # Measure transaction cost (simulated)
    # Blockchain: cost per transaction, Centralized: negligible cost
    blockchain_tx_cost = [10] * num_auth_attempts # Small, constant cost per block
    centralized_tx_cost = [0.01] * num_auth_attempts # Negligible

    # Measure computational overhead / scalability (simulated)
    # Scalability is inversly proportional to authentication time
    blockchain_scalability = [1 / lat for lat in blockchain_latencies]
    centralized_scalability = [1 / lat for lat in centralized_latencies]

    # Measure revocation propagation delay (simulated)
    # Blockchain: Depends on block time, Centralized: instant
    blockchain_revocation_delay = [10] * num_users # Assuming average 10 sec block time
    centralized_revocation_delay = [0.1] * num_users # Near-instant

    return {
        'blockchain_latencies': blockchain_latencies,
        'centralized_latencies': centralized_latencies,
        'blockchain_tx_cost': blockchain_tx_cost,
        'centralized_tx_cost': centralized_tx_cost,
        'blockchain_scalability': blockchain_scalability,
        'centralized_scalability': centralized_scalability,
        'blockchain_revocation_delay': blockchain_revocation_delay,
        'centralized_revocation_delay': centralized_revocation_delay
    }

# Run the simulation
results = run_performance_test(user_authentication, CentralizedSystem().authenticate)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_results(results):
    fig, axs = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Performance Comparison: BMFA vs. Centralized', fontsize=16)

    # Authentication Latency
    axs[0, 0].hist(results['blockchain_latencies'], bins=50, alpha=0.5, label='BMFA')
    axs[0, 0].hist(results['centralized_latencies'], bins=50, alpha=0.5, label='Centralized')
    axs[0, 0].set_title('Authentication Latency Distribution')
    axs[0, 0].set_xlabel('Time (s)')
    axs[0, 0].set_ylabel('Frequency')
    axs[0, 0].legend()

    # Transaction Cost
    axs[0, 1].bar(['BMFA', 'Centralized'], [np.mean(results['blockchain_tx_cost']), np.mean(results['centralized_tx_cost'])], color=['blue', 'orange'])
    axs[0, 1].set_title('Average Transaction Cost (Simulated)')
    axs[0, 1].set_ylabel('Cost per Auth')

    # Computational Overhead / Scalability
    axs[1, 0].bar(['BMFA', 'Centralized'], [np.mean(results['blockchain_scalability']), np.mean(results['centralized_scalability'])], color=['blue', 'orange'])
    axs[1, 0].set_title('Computational Overhead / Scalability (1/Latency)')
    axs[1, 0].set_ylabel('Throughput (1/s)')

    # Revocation Propagation Delay
    axs[1, 1].bar(['BMFA', 'Centralized'], [np.mean(results['blockchain_revocation_delay']), np.mean(results['centralized_revocation_delay'])], color=['blue', 'orange'])
    axs[1, 1].set_title('Average Revocation Propagation Delay')
    axs[1, 1].set_ylabel('Time (s)')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# Plot the results
plot_results(results)
```
